```python
import os
import warnings

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

```python
import zipfile

zip_path = "/content/UCI HAR Dataset.zip"
extract_path = "/content"

with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(extract_path)

print("Dataset extracted successfully!")
```

```python
import os

print(os.listdir("/content/UCI HAR Dataset"))
```

```python
dataset_path = "/content/UCI HAR Dataset"
```

```python
features = pd.read_csv(
    os.path.join(dataset_path, "features.txt"),
    sep=r"\s+",
    header=None,
    names=["Index", "Feature"]
)
```

```python
features.head()
```

```python
X_train = pd.read_csv(
    os.path.join(dataset_path, "train", "X_train.txt"),
    sep=r"\s+",
    header=None
)

X_train.columns = features["Feature"]

X_train.head()
```

```python
y_train = pd.read_csv(
    os.path.join(dataset_path, "train", "y_train.txt"),
    header=None,
    names=["Activity"]
)

y_train.head()
```

```python
subject_train = pd.read_csv(
    os.path.join(dataset_path, "train", "subject_train.txt"),
    header=None,
    names=["Subject"]
)

subject_train.head()
```

```python
train_df = pd.concat(
    [subject_train, y_train, X_train],
    axis=1
)

train_df.head()
```

```python
X_test = pd.read_csv(
    os.path.join(dataset_path, "test", "X_test.txt"),
    sep=r"\s+",
    header=None
)

X_test.columns = features["Feature"]

y_test = pd.read_csv(
    os.path.join(dataset_path, "test", "y_test.txt"),
    header=None,
    names=["Activity"]
)

subject_test = pd.read_csv(
    os.path.join(dataset_path, "test", "subject_test.txt"),
    header=None,
    names=["Subject"]
)

test_df = pd.concat(
    [subject_test, y_test, X_test],
    axis=1
)
```

```python
har_df = pd.concat(
    [train_df, test_df],
    ignore_index=True
)

har_df.head()
```

```python
print("Dataset Shape:", har_df.shape)
```

```python
har_df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
duplicates = har_df.duplicated().sum()

print("Duplicate rows:", duplicates)
```

```python
har_df.describe()
```

```python
activity_labels = pd.read_csv(
    os.path.join(dataset_path, "activity_labels.txt"),
    sep=r"\s+",
    header=None,
    names=["Activity", "Activity_Name"]
)

activity_labels
```

```python
activity_map = dict(zip(
    activity_labels["Activity"],
    activity_labels["Activity_Name"]
))

activity_map
```

```python
har_df["Activity_Name"] = har_df["Activity"].map(activity_map)
```

```python
har_df[["Activity", "Activity_Name"]].head(10)
```

```python
har_df["Activity_Name"].value_counts()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=har_df,
    x="Activity_Name",
    order=har_df["Activity_Name"].value_counts().index,
    palette="viridis"
)

plt.title("Distribution of Human Activities")
plt.xlabel("Activity")
plt.ylabel("Number of Samples")
plt.xticks(rotation=20)

plt.show()
```

```python
har_df["Subject"].value_counts().sort_index()
```

```python
plt.figure(figsize=(12,5))

sns.countplot(
    data=har_df,
    x="Subject",
    palette="viridis"
)

plt.title("Number of Samples per Subject")
plt.xlabel("Subject ID")
plt.ylabel("Number of Samples")

plt.show()
```

```python
selected_features = har_df.columns[2:12]

selected_features
```

```python
corr_matrix = har_df[selected_features].corr()

corr_matrix
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap (First 10 Features)")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    har_df["tBodyAcc-mean()-X"],
    bins=30,
    kde=True,
    color="steelblue"
)

plt.title("Distribution of tBodyAcc-mean()-X")
plt.xlabel("Acceleration Value")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x=har_df["tBodyAcc-mean()-X"],
    color="lightgreen"
)

plt.title("Box Plot of tBodyAcc-mean()-X")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=har_df,
    x="Activity_Name",
    y="tBodyAcc-mean()-X"
)

plt.title("Body Acceleration across Activities")
plt.xticks(rotation=30)

plt.show()
```

```python
feature_variance = har_df.iloc[:, 2:563].var()

top10 = feature_variance.sort_values(ascending=False).head(10)

top10
```

```python
plt.figure(figsize=(12,6))

top10.plot(kind="bar")

plt.title("Top 10 Features with Highest Variance")
plt.xlabel("Feature")
plt.ylabel("Variance")

plt.xticks(rotation=90)

plt.show()
```
