# Human Activity Recognition using Machine Learning

## Objective

The objective of this project is to develop a machine learning system capable of recognizing human activities using smartphone sensor data. The model will classify activities such as walking, sitting, standing, laying, walking upstairs, and walking downstairs based on accelerometer and gyroscope signals.

```python
import os
import warnings

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

```python
import zipfile

zip_path = "/content/UCI HAR Dataset.zip"
extract_path = "/content"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Dataset extracted successfully!")
```

```python
dataset_path = "/content/UCI HAR Dataset"
```

```python
features = pd.read_csv(
    os.path.join(dataset_path, "features.txt"),
    sep=r"\s+",
    header=None,
    names=["Index", "Feature"]
)
```

```python
features.head()
```

```python
X_train = pd.read_csv(
    os.path.join(dataset_path, "train", "X_train.txt"),
    sep=r"\s+",
    header=None
)
```

```python
X_train.shape
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_train.columns = features["Feature"]
```

```python
X_train.head()
```

```python
print("Shape of X_train:", X_train.shape)
```

```python
X_train.columns[:10]
```

```python
y_train = pd.read_csv(
    os.path.join(dataset_path, "train", "y_train.txt"),
    header=None,
    names=["Activity"]
)
```

```python
y_train.head()
```

```python
y_train.shape
```

```python
subject_train = pd.read_csv(
    os.path.join(dataset_path, "train", "subject_train.txt"),
    header=None,
    names=["Subject"]
)
```

```python
subject_train.head()
```

```python
subject_train.shape
```

```python
train_df = pd.concat([subject_train, y_train, X_train], axis=1)
```

```python
train_df.head()
```

```python
train_df.shape
```

```python
X_test = pd.read_csv(
    os.path.join(dataset_path, "test", "X_test.txt"),
    sep=r"\s+",
    header=None
)

X_test.columns = features["Feature"]
```

```python
y_test = pd.read_csv(
    os.path.join(dataset_path, "test", "y_test.txt"),
    header=None,
    names=["Activity"]
)
```

```python
subject_test = pd.read_csv(
    os.path.join(dataset_path, "test", "subject_test.txt"),
    header=None,
    names=["Subject"]
)
```

```python
test_df = pd.concat(
    [subject_test, y_test, X_test],
    axis=1
)
```

```python
test_df.head()
```

```python
test_df.shape
```

```python
har_df = pd.concat(
    [train_df, test_df],
    axis=0,
    ignore_index=True
)
```
